```python
# Import the packages 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

# Read the data
df = pd.read_csv(r'C:\Users\fikru\Downloads\mon\movies.csv')
```

```python
#View the First Few Rows

df.head()
```

```python
# Summary Statistics

df.describe()
```

```python
# Missing Values

df.isnull().sum()
```

```python
# Duplicate Rows

df.duplicated().sum()
```

```python
# Data Types

df.dtypes
```

```python
# Remove duplicates 

df.drop_duplicates()
```

```python
# Sorting by a Column

df_sorted = df.sort_values('gross', ascending=False)
df_sorted.head()
```

```python
# Filtering on rating above 8

df_filtered = df[df['score'] > 8.0]
df_filtered.head()
```

```python
# Filling Missing Values

df['budget'].fillna(df['budget'].mean(), inplace=True)
df['gross'].fillna(df['gross'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# Creating a New Column

df['profit'] = df['gross'] - df['budget']
```

```python
# Removing Leading and Trailing Whitespaces

df['genre'] = df['genre'].str.strip()
```

```python
# Converting to Uppercase

df['rating'] = df['rating'].str.upper()
```

```python
# Pearson Correlation

df.corr(method='pearson', numeric_only=True)
```

```python
# Spearman Correlation

df.corr(method='spearman', numeric_only=True)
```

```python
# Kendall Correlation

df.corr(method='kendall', numeric_only=True)
```

```python
# Heatmap - Correlation Matrix

correlation_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Correlation between gross and budget 

df['budget'].corr(df['gross'])
```

```python
# Correlation above a specified threshold value 

threshold = 0.5
high_correlation = df.corr(numeric_only=True).abs().stack().reset_index().rename(columns={0: 'correlation'})
high_correlation = high_correlation[high_correlation['correlation'] > threshold]
high_correlation
```

```python
# Correlation on gross and budget

df.plot.scatter(x='gross', y='budget')
```

```python
# Scatter plots for all numeric columns

pd.plotting.scatter_matrix(df, figsize=(10, 10))
plt.show()
```

```python
# Bar Plot - Genre Distribution

genre_counts = df['genre'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Genre Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram - Rating Distribution

plt.figure(figsize=(8, 6))
plt.hist(df['score'], bins=10, edgecolor='black')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Rating Distribution')
plt.show()
```

```python
# Scatter Plot - Budget vs. Gross Revenue

plt.figure(figsize=(8, 6))
plt.scatter(df['budget'], df['gross'], color='green')
plt.xlabel('Budget')
plt.ylabel('Gross Revenue')
plt.title('Budget vs. Gross Revenue')
plt.show()
```

```python
# Box Plot - Runtime Distribution by Genre

plt.figure(figsize=(10, 6))
sns.boxplot(x='genre', y='runtime', data=df)
plt.xlabel('Genre')
plt.ylabel('Runtime')
plt.title('Runtime Distribution by Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
# Line Plot - Average Rating Over Years

avg_rating_year = df.groupby('year')['score'].mean()
plt.figure(figsize=(10, 6))
plt.plot(avg_rating_year.index, avg_rating_year.values)
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.title('Average Rating Over Years')
plt.show()
```

```python
# Pie Chart - Country Distribution

country_counts = df['country'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%')
plt.title('Country Distribution')
plt.show()
```
